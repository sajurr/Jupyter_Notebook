import matplotlib.pyplot as plt
import numpy as np

```python
theta = np.linspace(0, 2*np.pi, 1000)
r = 1 + 3/4 * np.sin(3*theta)
x = r * np.cos(theta)
y = r * np.sin(theta)
plt.plot(x, y)
```

```python
Area = 1/2 * sum(r**2) * (theta[1] - theta[0])
print(Area)
```

```python
Arc_length = sum(np.sqrt(r**2 + np.gradient(r, theta)**2)) * (theta[1] - theta[0])
print(Arc_length)
```

$$\frac{P}{(\sigma \epsilon {T_0}^{4})} = \left(\frac{1}{1 + e^{-kt}}\right)^4$$

```python
kt = np.linspace(0, 3, 100)
Power_dimensionless = (1/(1+np.exp(-kt)))**4
plt.plot(kt, Power_dimensionless)
E = np.cumsum((Power_dimensionless) * (kt[1]-kt[0]))
plt.plot(kt, E)
plt.xlabel('$kt$', fontsize=20)
plt.ylabel(r'$\left(\frac{k}{A \sigma \epsilon {T_0}^{4}}\right) E(k)$', fontsize=20)
```
