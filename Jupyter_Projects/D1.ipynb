```python
x = np.linspace(1, 10, 100)
y = 1/x**2 * np.sin(x)
dydx = np.gradient(y, x)
y_int = np.cumsum(y) * (x[1]-x[0])
plt.plot(x, y)
plt.plot(x, dydx)
plt.plot(x, y_int)
```

```python
N = 10000
x = np.linspace(0, 10, N + 1)
y = np.exp(-x/10) * np.sin(x)
dydx = np.gradient(y, x)
d2ydx2 = np.gradient(dydx, x)
plt.plot(x, y)
plt.plot(x, dydx)
plt.plot(x, d2ydx2)
```

```python
theta = np.linspace(0, 2*np.pi, 1000)
r = 1 + 3/4 * np.sin(3*theta)
x = r * np.cos(theta)
y = r * np.sin(theta)
plt.plot(x, y)
```

```python
Area = 1/2 * sum(r**2) * (theta[1] - theta[0])
print(Area)
```

```python
Arc_length = sum(np.sqrt(r**2 + np.gradient(r, theta)**2)) * (theta[1] - theta[0])
print(Arc_length)
```
